# Commands
## Navigation
- ls: list all files and directories in  current working directory.  
ls Documents/  -> Relative path  
ls /home/joe/Documents/ -> Absoulte path
ls -a -> list all files including hidden ones  
ls -l -> including file information
<br/><br/>
- pwd: print your current working directory (absoulte path).
<br/><br/>
- cd: change directory  
cd (without any arguemnt) / cd ~: go to home  
<br/><br/>
- ..: move up one level up (to parent directory).
<br/><br/>
- pushd / popd: used to go to where I worked.
<br/><br/>
- file: print the extenstion of a file.
file .face
<br/><br/>
- locate: find a file.  
locate fstab  
<br/><br/>
- which: search the list of paths
which cal
<br/><br/>
- history: print all the previous executed commands.

## Getting help
- whatis: assist in getting a short explanation of any command line.  
whatis cal
<br/><br/>
- apropos: helps the user when they don't remember the exact command but knows a few keywords related to the command that define its uses or functionality.  
appropos time
<br/><br/>
- man: display the user manual of any command that we can run on the terminal.  
man man

## Working with files
- mkdir: make a directory.  
mkdir Junk Junk2 
<br/><br/>
- touch: change file timestamps/create empty files
touch file1 file2 file3
<br/><br/>
- cp: copy.  
cp ~/.bashrc bashrc
<br/><br/>
- mv: move or rename.  
mv bashrc.bak bashrc
<br/><br/>
- rm: remove.  
rm *: remove all files in the directory.  
rm file*: remove files starting with file  
rm -r / rmdir: remove directory

## Text files
- cat: print the content of a file onto the standard output stream.  
cat >> file2: append text  
cat > file2: new one  
cat file1 fil2
<br/><br/>
- more: view the text files in the command prompt, displaying one screen at a time in case the file is large.  
more file1
<br/><br/>
- less: show a file's contents one screen at a time..  
less file1
history | less
<br/><br/>
- nano: use a text editor.  
nano file2
<br/><br/>
- ls -al / > all.txt

## Users
- sudo: allows programs to be executed as a super user (aka root user) or another user.  
sudo -s: be a root user
<br/><br/>
- su: change user ID or become another user.  
su - cindy
<br/><br/>
- users: print the user names of users currently logged in.
<br/><br/>
- id: print real and effective user and group IDs

## Change file permissions
- chmod: change permissions of a file  
chmod +x file1  
chmod 700 file1


## Killing programs and logging out
- Ctrl+C: kill a running command
- killall: kill processes by name  
kill firefox
<br/><br/>
- exit: log out of bash

## Useful shortcuts
- Ctrl+D: signal bash that there is no more input
- Ctrl+L: redraw the screen
- Ctrl++: make text bigger in terminal emulator
- Ctrl+-: make text smaller in terminal emulator

# Tip
- 'tab' key is used for autocomplete.
